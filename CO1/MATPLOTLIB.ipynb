```python

import matplotlib.pyplot as plt
import numpy as np
x=np.linspace(0,5,11)
y=x**2
x
```

```python
y
```

```python
plt.plot(x,y,'r')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('graph')
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r--')
plt.subplot(1,2,2)
plt.plot(x,y,'b*-')
```

```python
plt.subplot(1,3,1)
plt.plot(x,y,'g--')
plt.subplot(1,3,2)
plt.plot(x,y,'r*-')
plt.subplot(1,3,3)
plt.plot(x,y,'b--')
```

```python

fig=plt.figure()
axes=fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y,'b')
axes.set_xlabel('x axis')
axes.set_ylabel('y axis')
axes.set_title('graph')
```

```python
fig=plt.figure()
axes1=fig.add_axes([0.1,0.1,0.8,0.8])
axes2=fig.add_axes([0.2,0.4,0.2,0.3])

axes1.plot(x,y,'b')
axes1.set_xlabel('x axis')
axes1.set_ylabel('y axis')
axes1.set_title('graph')

axes2.plot(x,y,'r')
axes2.set_xlabel('x axis')
axes2.set_ylabel('y axis')
axes2.set_title('graph')

```

```python
fig,axes=plt.subplots(1,2)
axes
```

```python
fig,axes=plt.subplots(1,2)
for ax in axes:
  ax.plot(x,y,'r')
  ax.set_xlabel('x axis')
  ax.set_ylabel('y axis')
  ax.set_title('graph')


```
